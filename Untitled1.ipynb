```python
%matplotlib inline
import swagger_client 
import os

access_key, secret_key =  os.environ["BITMEX_ACCESS_KEY"], os.environ["BITMEX_ACCESS_SECRET_KEY"]

from __future__ import print_function
import time
import swagger_client
from swagger_client.rest import ApiException
from pprint import pprint

# Configure API key authorization: apiKey
configuration = swagger_client.Configuration()
configuration.api_key['api-key'] = secret_key
configuration.api_key['api-nonce'] = secret_key
configuration.api_key['api-signature'] = secret_key
configuration.host = 'https://www.bitmex.com/api/v1'
configuration.proxy = 'http://localhost:1087'
apiClient = swagger_client.ApiClient(configuration)
# create an instance of the API class
api_instance = swagger_client.APIKeyApi(apiClient)


symbol = 'XBTUSD'

instrument_instance = swagger_client.InstrumentApi(apiClient)

orderBook_instance = swagger_client.OrderBookApi(apiClient)

orderBook = orderBook_instance.order_book_get_l2(symbol=symbol, depth=10)

import pandas as pd

def transform(ob, size=10):
    if len(ob) != size*2:
        raise Exception('Wrong size of order book')
    asks = ob[:size][::-1]
    bids = ob[size:]
    return asks, bids

import time
obs = []
cnt = 0
for i in range(100000):
    try:
        ob = orderBook_instance.order_book_get_l2(symbol=symbol, depth=10)
        asks, bids = transform(ob)
        row = {}
        for index, val in enumerate(asks):
            ind = str(index)
            row['ask_p_' + ind] = val.price
            row['ask_vol_' + ind] = val.size
        for index, val in enumerate(bids):
            ind = str(index)
            row['bid_p_' + ind] = val.price
            row['bid_vol_' + ind] = val.size
        obs.append(row)
    except ApiException:
        pprint("Something bad happened...")
        cnt += 1
    finally:
        if cnt > 5:
            break
        time.sleep(1)

df = pd.DataFrame(obs)

df.describe()
df.to_csv('./orderbooks.csv')
```

# bid ask spreads and mid prices(n levels)

```python
for i in range(10):
    ind = str(i)
    df['spreed_'+ind] = df['ask_p_'+ind] - df['bid_p_'+ind]
    df['mid_p'+ind] = (df['ask_p_'+ind] + df['bid_p_'+ind]) / 2
```

# price difference

```python
for i in range(9):
    ind = str(i+1)
    df['ask_p_diff_'+ind+'_0'] = df['ask_p_'+ind] - df['ask_p_0']
    df['bid_p_diff_'+ind+'_0'] = df['bid_p_'+ind] - df['bid_p_0']
    ind1 = str(i)
    df['ask_p_diff_'+ind+'_'+ind1] = df['ask_p_'+ind] - df['ask_p_'+ind1]
    df['bid_p_diff_'+ind+'_'+ind1] = df['bid_p_'+ind] - df['bid_p_'+ind1]
```

# mean prices and volumes

```python
l1, l2, l3, l4 = [], [], [], []
for i in range(10):
    ind = str(i)
    l1.append('ask_p_'+ind)
    l2.append('bid_p_'+ind)
    l3.append('ask_vol_'+ind)
    l4.append('bid_vol_'+ind)
df['ask_p_mean'] = df[l1].mean(axis=1)
df['bid_mean'] = df[l2].mean(axis=1)
df['ask_vol_mean'] = df[l3].mean(axis=1)
df['bid_vol_mean'] = df[l4].mean(axis=1)
```

```python
for i in range(10):
    ind = str(i)
    df['spreed_vol_'+ind] = df['ask_vol_'+ind] - df['bid_vol_'+ind]
l = []
template = 'spreed_vol_{0}'
for i in range(10):
    l.append(template.format(i))
df['accum_spreed_vol'] = df[l].sum(axis=1)
df.head()
```
